```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Load the AIA 304 image (you need to provide the path to your image)
image = cv2.imread('img4.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Use Canny edge detection to detect edges (likely regions of plasma ejection)
edges = cv2.Canny(blurred, 50, 150)

# Use a threshold to highlight bright regions (potential plasma ejections)
_, thresholded = cv2.threshold(image, 200, 255, cv2.THRESH_BINARY)

# Combine edges and thresholded image for better plasma detection
combined = cv2.bitwise_or(edges, thresholded)

# Find contours (areas of plasma ejection)
contours, _ = cv2.findContours(combined, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image (optional)
output_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(output_image, contours, -1, (0, 255, 0), 2)

# Display the results using matplotlib
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original AIA 304 Image')

plt.subplot(1, 2, 2)
plt.imshow(output_image)
plt.title('Plasma Ejection Contours')

plt.show()

# Save the images
cv2.imwrite('original_image.png', image)
cv2.imwrite('plasma_contours.png', output_image)


```
